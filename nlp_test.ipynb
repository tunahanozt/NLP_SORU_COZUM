MODEL HAZIRLAMA

```python
from llama_cpp import Llama
from huggingface_hub import hf_hub_download
import os

REPO_ID = "Garoz/matematik_soru_cozumu"
FILENAME = "gemma-final-tuna.gguf"

print("📥 Model indiriliyor...")
model_path = hf_hub_download(repo_id=REPO_ID, filename=FILENAME, token=False)

print("🚀 Model GPU belleğine yükleniyor...")
llm = Llama(
    model_path=model_path,
    n_ctx=2048,        # Bağlam penceresini genişlettik
    n_gpu_layers=-1,   # Tam GPU desteği
    n_batch=512,
    verbose=False
)

print("✅ Model kullanıma hazır!")
```

TEST FONKSİYONU

```python
test_sorusu = "Bir sepette 12 elma var. 3 tanesini yedim, 4 tanesini arkadaşıma verdim. Sepette kaç elma kaldı?"

prompt_format = f"""<start_of_turn>user
Soru: {test_sorusu}<end_of_turn>
<start_of_turn>model
"""

output = llm(
    prompt_format,
    max_tokens=256,
    stop=["<end_of_turn>", "<start_of_turn>", "user", "####"], 
    temperature=0.1,
    echo=False
)

print(output['choices'][0]['text'])
```
