MODEL HAZIRLAMA

```python
from llama_cpp import Llama
import os

# İndirdiğin .gguf dosyasının bilgisayarındaki tam yolu
MODEL_PATH = "gemma-final-tuna.gguf"

if os.path.exists(MODEL_PATH):
    print("✅ Model dosyası bulundu, yükleniyor...")
    # n_gpu_layers: GPU varsa -1 yap, yoksa 0 yap.
    # n_ctx: Hafıza uzunluğu (Matematik soruları için 512-1024 yeterli)
    llm = Llama(
        model_path=MODEL_PATH,
        n_ctx=1024,
        n_gpu_layers=-1 if "CUDA" in os.environ.get("PATH", "") else 0,
        verbose=False # Log kalabalığını gizlemek için
    )
    print("🚀 Model hazır!")
else:
    print("❌ HATA: Model dosyası belirtilen yolda bulunamadı!")
```

TEST FONKSİYONU

```python
def matematik_cozucu(soru):
    # Senin eğitimde kullandığın ve modelin en iyi anladığı format
    prompt = f"<bos><start_of_turn>user\nSoru: {soru}\n<end_of_turn>\n<start_of_turn>model\nCevap:"
    
    response = llm(
        prompt,
        max_tokens=512,
        stop=["<end_of_turn>", "Soru:"],
        temperature=0.2, # Daha kararlı ve kesin cevaplar için düşük tuttuk
        echo=False
    )
    
    return response['choices'][0]['text']

##SORU BURAYA YAZILACAK.
test_sorusu = "Bir sepette 12 elma var. 3 tanesini yedim, 4 tanesini arkadaşıma verdim. Kaç elma kaldı?"
cozum = matematik_cozucu(test_sorusu)

print(f"SORU: {test_sorusu}")
print("-" * 30)
print(f"MODELİN ÇÖZÜMÜ:\n{cozum}")
```
